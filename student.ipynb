**4. DATA VISUALIZATION**

**4.1 Vertical bar graph showing profitable movie genres by summing up the domestic and foreign gross from 2015-2018**

```python
# Convert domestic_gross and foreign_gross columns to numeric
merged_data['domestic_gross'] = pd.to_numeric(merged_data['domestic_gross'], errors='coerce')
merged_data['foreign_gross'] = pd.to_numeric(merged_data['foreign_gross'], errors='coerce')
```

```python
#find the maximum year in my merged_data
largest_year = merged_data['year'].max()
print("The largest year in the dataset is:", largest_year)
```

```python
recent_years_data = merged_data[(merged_data['year'] >= 2015) & (merged_data['year'] <= 2018)]

# Calculating the total gross income
recent_years_data['total_gross'] = recent_years_data['domestic_gross'] + recent_years_data['foreign_gross']

# Group by genre and calculate the total gross income for each genre
genre_total_gross = recent_years_data.groupby('genres')['total_gross'].sum().reset_index()

# Sorting genres by total gross income in descending order
top_genres = genre_total_gross.sort_values(by='total_gross', ascending=False).head(10)

plt.figure(figsize=(10, 10))  

sns.barplot(x='genres', y='total_gross', data=top_genres, color='blue')  
plt.title("Top-Performing Genres at the Box Office (2015-2018)")
plt.xlabel("Genre")
plt.ylabel("Total Gross Income (in billions)")
#rotating labels for readability
plt.xticks(rotation=90)

plt.show()
```

**4.2 Pie-chart showing the Genre Distribution from 2015-2018**

```python
# Filter the data for recent years (2015-2018)
recent_years_data = merged_data[(merged_data['year'] >= 2015) & (merged_data['year'] <= 2018)]

# Calculate genre counts for recent years
genre_counts = recent_years_data['genres'].str.split(',').explode().value_counts()

# Selecting the top 8 genres
top_genres = genre_counts.head(8)

plt.figure(figsize=(8, 8))
plt.pie(top_genres, labels=top_genres.index, autopct='%1.1f%%', startangle=140)
plt.title("Genre Distribution (Top 8 Genres) - 2015 to 2018")
# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')
plt.show()
```

**4.3 horizontal bar graph showing the leading studio using the total gross from 2015-2018**

```python
recent_years_data = merged_data[(merged_data['year'] >= 2015) & (merged_data['year'] <= 2018)]

merged_data['total_gross'] = merged_data['domestic_gross'] + merged_data['foreign_gross']

# Group the data by studio and calculate the total gross income for each studio
studio_gross = merged_data.groupby('studio')['total_gross'].sum().reset_index()

# Find the leading N studios with the highest total gross income
top_n_studios = 10  # Change this to the desired number of top studios
leading_studios = studio_gross.nlargest(top_n_studios, 'total_gross')

plt.figure(figsize=(10, 6))


sns.barplot(x='total_gross', y='studio', data=leading_studios, color='pink')

plt.title("Top {} Studios by Total Gross Income".format(top_n_studios))
plt.xlabel("Total Gross Income")
plt.ylabel("Studio")

plt.show()
```

**4.4 horizontal bar graph showing the average rating of the first 10 genres from 2015-2018**

```python
recent_years_data = merged_data[(merged_data['year'] >= 2015) & (merged_data['year'] <= 2018)]

# Calculate genre average ratings for recent years
genre_avg_ratings = recent_years_data.groupby('genres')['averagerating'].mean().reset_index()

# Sort the genres by average rating in descending order and select the top 10
top_10_genres = genre_avg_ratings.nlargest(10, 'averagerating')

plt.figure(figsize=(10, 6))

sns.barplot(x='averagerating', y='genres', data=top_10_genres, color='purple')
plt.title("Average Rating of the Top 10 Genres (2015-2018)")
plt.xlabel("Average Rating")
plt.ylabel("Genre")

plt.show()
```

**4.5 horizontal bar graph showing the num-votes of the first 10 genres from 2015-2018**

```python
recent_years_data = merged_data[(merged_data['year'] >= 2015) & (merged_data['year'] <= 2018)]

# Calculating the total number of votes for each genre
genre_num_votes = recent_years_data.groupby('genres')['numvotes'].sum().reset_index()

# Sorting the genres by total number of votes in descending order
genre_num_votes = genre_num_votes.sort_values(by='numvotes', ascending=False)

# Select the top 10 genres
top_10_genres = genre_num_votes.head(10)

plt.figure(figsize=(10, 8))

sns.barplot(x='numvotes', y='genres', data=top_10_genres, color='green')
plt.title("Total Number of Votes by Genre (2015-2018)")
plt.xlabel("Total Number of Votes")
plt.ylabel("Genre")
plt.show()
```
